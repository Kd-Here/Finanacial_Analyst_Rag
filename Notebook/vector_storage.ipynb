# Creating Faiss DB
[Faiss Python Docs](https://github.com/facebookresearch/faiss/wiki/Getting-started)

```python
import faiss
dir(faiss)
```

```python
index  = faiss.IndexFlatL2(vector_dim)   # build the index

type(index),dir(index)
```

```python
vectors.shape
```

```python
index.ntotal
```

```python
index.add(vectors)
```

```python
index,index.d
```

```python
search_query = 'I want to eat an apple'
search_vector = encoder.encode(search_query)
print(type(search_vector))
print(search_vector.shape)
print(search_vector)
```

```python
import numpy as np

# 1. Reshape the vector to be 2D (1 row, 384 columns)
search_vector_2d = search_vector.reshape(1, -1)
print(search_vector_2d.shape)  # Output: (1, 384)

# 2. Unpack BOTH outputs from the search function
A = index.search(x=search_vector_2d, k=3)
print(A,type(A))
```

```python
# error faiss requires 2d array or matrx to search
# topk_nearest_vectors_search_vector = index.search(x=search_vector,k=1)

import numpy as np

# 1. Reshape the vector to be 2D (1 row, 384 columns)
search_vector_2d = search_vector.reshape(1, -1)
print(search_vector_2d.shape)  # Output: (1, 384)

# 2. Unpack BOTH outputs from the search function
distances, indices = index.search(x=search_vector_2d, k=3)
print(type(distances),type(indices))


# 3. View your results
print("Nearest neighbor IDs:", indices)
# print(indices[0][0])
print("Squared L2 distances:", distances)

for i in indices[0]:
    print("Index:- ",i,"Corresponding text in datafram:- ",df.text[i])
```
