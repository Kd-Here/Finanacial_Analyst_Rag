```python


from langchain_text_splitters import CharacterTextSplitter

chunks = CharacterTextSplitter(separator='\n',is_separator_regex=True,chunk_size=300)
chunks = chunks.split_text(random_text)
type(chunks),len(chunks)
```

```python
for i in chunks:
    print(i)
    print("\n")
```

```python
import langchain_text_splitters

from langchain_text_splitters import RecursiveCharacterTextSplitter
recir_chunks = RecursiveCharacterTextSplitter(separators=['\n\n','\n','.'],chunk_size=300)

recir_chunks = recir_chunks.split_text(random_text)
type(recir_chunks),len(recir_chunks)
```
